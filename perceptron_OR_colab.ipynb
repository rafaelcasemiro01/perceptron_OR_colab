## 📋 Célula 2 — Tabelas Verdade
Definimos as entradas e saídas para AND e OR.
**A única diferença está no vetor `saidas_OR`.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Entradas: todas as combinações possíveis de 2 bits
entradas = np.array([
    [0, 0],   # linha 1
    [0, 1],   # linha 2
    [1, 0],   # linha 3
    [1, 1],   # linha 4
])

# Saídas AND: só é 1 quando AMBAS as entradas são 1
saidas_AND = np.array([0, 0, 0, 1])

# Saídas OR:  é 1 quando PELO MENOS UMA entrada é 1  ← única mudança!
saidas_OR  = np.array([0, 1, 1, 1])

# Visualiza as duas tabelas lado a lado
print('Tabela Verdade')
print(f'{"X1":>6} {"X2":>6} {"AND":>6} {"OR":>6}')
print('-' * 28)
for i, (xi, ya, yo) in enumerate(zip(entradas, saidas_AND, saidas_OR)):
    diff = '<- OR = 1' if ya != yo else ''
    print(f'{xi[0]:>6} {xi[1]:>6} {ya:>6} {yo:>6}  {diff}')
```

## 🤖 Célula 3 — Classe Perceptron
Implementação completa do Perceptron com:
- Inicialização dos pesos em zero
- Função de ativação degrau
- Regra de aprendizado de Rosenblatt
- Histórico de erros e pesos para visualização

```python
class Perceptron:
    """
    Perceptron simples de Rosenblatt (1958).
    Aprende ajustando pesos quando erra uma previsão.
    """

    def __init__(self, taxa_aprendizado=0.1, epocas=10):
        self.taxa   = taxa_aprendizado  # o quanto corrige a cada erro
        self.epocas = epocas            # número máximo de passagens pelos dados
        self.pesos  = None              # serão inicializados no treino
        self.bias   = None
        self.historico_erros = []       # erros por época (para gráfico)
        self.historico_pesos = []       # pesos por época  (para gráfico)

    def ativar(self, soma):
        """Função degrau: retorna 1 se soma >= 0, senão 0."""
        return 1 if soma >= 0 else 0

    def treinar(self, X, y, verbose=True):
        """Treina o Perceptron nos dados X com rótulos y."""
        # Inicializa pesos e bias em zero
        self.pesos = np.zeros(X.shape[1])  # um peso por entrada
        self.bias  = 0.0
        self.historico_erros = []
        self.historico_pesos = []

        for epoca in range(1, self.epocas + 1):
            total_erros = 0

            if verbose:
                print(f'\n{'='*55}')
                print(f'  ÉPOCA {epoca}')
                print(f'{'='*55}')

            for i, (xi, yi) in enumerate(zip(X, y)):
                # 1. Calcula a saída atual
                soma  = np.dot(xi, self.pesos) + self.bias
                saida = self.ativar(soma)

                # 2. Calcula o erro
                erro = yi - saida
                total_erros += abs(erro)

                # 3. Atualiza pesos e bias se houver erro
                self.pesos += self.taxa * erro * xi
                self.bias  += self.taxa * erro

                if verbose:
                    sinal = '+' if erro >= 0 else ''
                    print(f'  Amostra {i+1}: entrada={xi}'
                          f' | esperado={yi} | obtido={saida}'
                          f' | erro={sinal}{erro}'
                          f' | pesos={np.round(self.pesos,2)}'
                          f' | bias={self.bias:.2f}')

            # Salva histórico
            self.historico_erros.append(total_erros)
            self.historico_pesos.append(self.pesos.copy())

            if verbose:
                print(f'\n  >> Total de erros na época {epoca}: {total_erros}')

            # Para se não houve nenhum erro
            if total_erros == 0:
                if verbose:
                    print(f'  >> ✅ Convergiu na época {epoca}!')
                break

    def prever(self, X):
        """Faz previsões para um conjunto de entradas."""
        return [self.ativar(np.dot(xi, self.pesos) + self.bias) for xi in X]

print('✅ Classe Perceptron definida!')
```

## 🏋️ Célula 4 — Treinamento com AND
Executamos o treinamento com a tabela AND para ter uma **base de comparação**.
Você verá cada ajuste de peso impresso na tela.

```python
print('TREINAMENTO — AND')
print('(base de comparação)\n')

p_and = Perceptron(taxa_aprendizado=0.1, epocas=10)
p_and.treinar(entradas, saidas_AND, verbose=True)

# Resultado final
previsoes_and = p_and.prever(entradas)
acertos = sum(yi == pi for yi, pi in zip(saidas_AND, previsoes_and))

print('\n' + '-'*55)
print('RESULTADO FINAL — AND')
print('-'*55)
print(f'Pesos finais : {p_and.pesos}')
print(f'Bias final   : {p_and.bias}')
print(f'Épocas usadas: {len(p_and.historico_erros)}')
print(f'Acurácia     : {acertos}/{len(saidas_AND)} ({acertos/len(saidas_AND)*100:.0f}%)')
```

## 🔄 Célula 5 — Treinamento com OR
Agora trocamos apenas o vetor de saídas para OR.
**Observe como os pesos e bias finais são diferentes do AND.**

```python
print('TREINAMENTO — OR')
print('(tabela trocada: saidas_OR ao invés de saidas_AND)\n')

p_or = Perceptron(taxa_aprendizado=0.1, epocas=10)
p_or.treinar(entradas, saidas_OR, verbose=True)

# Resultado final
previsoes_or = p_or.prever(entradas)
acertos_or = sum(yi == pi for yi, pi in zip(saidas_OR, previsoes_or))

print('\n' + '-'*55)
print('RESULTADO FINAL — OR')
print('-'*55)
print(f'Pesos finais : {p_or.pesos}')
print(f'Bias final   : {p_or.bias}')
print(f'Épocas usadas: {len(p_or.historico_erros)}')
print(f'Acurácia     : {acertos_or}/{len(saidas_OR)} ({acertos_or/len(saidas_OR)*100:.0f}%)')
```

## 📊 Célula 6 — Comparação AND × OR
Tabela lado a lado mostrando o resultado de cada amostra nos dois modelos.

```python
print('COMPARAÇÃO FINAL — AND × OR')
print(f'{"Entrada":<12} {"Esp AND":>9} {"Prev AND":>10}'
      f' {"Esp OR":>8} {"Prev OR":>9} {"OR ok?":>7}')
print('-' * 62)

for xi, ya, pa, yo, po in zip(entradas, saidas_AND, previsoes_and,
                               saidas_OR,  previsoes_or):
    ok = '✅' if yo == po else '❌'
    print(f'{str(xi):<12} {ya:>9} {pa:>10} {yo:>8} {po:>9} {ok:>7}')

print()
print(f'AND — Pesos: {p_and.pesos}  Bias: {p_and.bias}')
print(f'OR  — Pesos: {p_or.pesos}   Bias: {p_or.bias}')
```

## 📈 Célula 7 — Gráficos
Três visualizações:
- **Esquerda:** curva de erros por época (AND × OR)
- **Centro:** fronteira de decisão aprendida para o OR
- **Direita:** como os pesos evoluíram durante o treinamento do OR

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4.5))
fig.patch.set_facecolor('white')

AZUL  = '#0071E3'
CINZA = '#6E6E73'
FUNDO = 'white'

def formatar_eixo(ax):
    ax.set_facecolor(FUNDO)
    for lado in ['top', 'right']:
        ax.spines[lado].set_visible(False)
    for lado in ['left', 'bottom']:
        ax.spines[lado].set_color('#D2D2D7')
        ax.spines[lado].set_linewidth(0.6)
    ax.tick_params(colors=CINZA, labelsize=8, length=0)
    ax.grid(axis='y', color='#D2D2D7', linewidth=0.5, alpha=0.8)

# ── Gráfico 1: Curva de erros ──────────────────────────
ax = axes[0]
formatar_eixo(ax)

ax.plot(range(1, len(p_and.historico_erros)+1), p_and.historico_erros,
        'o-', color=CINZA, lw=1.5, ms=6, mfc='white', mew=1.8, label='AND')
ax.plot(range(1, len(p_or.historico_erros)+1),  p_or.historico_erros,
        'o-', color=AZUL,  lw=1.5, ms=6, mfc='white', mew=1.8, label='OR')

ax.set_title('Erro por Época', color='#1D1D1F', fontsize=11,
             fontweight='bold', loc='left', pad=10)
ax.set_xlabel('Época', color=CINZA, fontsize=8.5)
ax.set_ylabel('Número de erros', color=CINZA, fontsize=8.5)
ax.set_xticks([1, 2, 3, 4])
ax.legend(frameon=False, labelcolor=CINZA, fontsize=8)

# ── Gráfico 2: Fronteira de decisão — OR ──────────────
ax = axes[1]
formatar_eixo(ax)
ax.grid(True, color='#D2D2D7', linewidth=0.5, alpha=0.8)  # grade completa aqui

cores = {0: '#D2D2D7', 1: AZUL}
for xi, yi in zip(entradas, saidas_OR):
    ax.scatter(xi[0], xi[1], c=cores[yi], s=110, zorder=5,
               edgecolors='white', linewidths=2)
    ax.annotate(f'({xi[0]},{xi[1]})→{yi}',
                xy=(xi[0], xi[1]), xytext=(xi[0]+0.07, xi[1]+0.06),
                color=CINZA, fontsize=8)

# Desenha a fronteira de decisão: w1*x1 + w2*x2 + b = 0
w1, w2, b = p_or.pesos[0], p_or.pesos[1], p_or.bias
if abs(w2) > 1e-9:
    x_vals = np.linspace(-0.35, 1.35, 300)
    y_vals = -(w1 * x_vals + b) / w2
    ax.plot(x_vals, y_vals, color='#D2D2D7', lw=1.2, ls='--')

ax.set_xlim(-0.3, 1.3)
ax.set_ylim(-0.3, 1.3)
ax.set_title('Fronteira de Decisão — OR', color='#1D1D1F',
             fontsize=11, fontweight='bold', loc='left', pad=10)
ax.set_xlabel('Entrada X1', color=CINZA, fontsize=8.5)
ax.set_ylabel('Entrada X2', color=CINZA, fontsize=8.5)

# ── Gráfico 3: Evolução dos pesos — OR ────────────────
ax = axes[2]
formatar_eixo(ax)

epocas_or = range(1, len(p_or.historico_pesos)+1)
pesos_hist = np.array(p_or.historico_pesos)

ax.plot(epocas_or, pesos_hist[:, 0], 'o-', color=AZUL,  lw=1.5,
        ms=6, mfc='white', mew=1.8, label='W1')
ax.plot(epocas_or, pesos_hist[:, 1], 'o--', color=CINZA, lw=1.5,
        ms=6, mfc='white', mew=1.8, label='W2')
ax.axhline(y=p_or.bias, ls=':', color='#D2D2D7', lw=1, label='Bias')

ax.set_title('Evolução dos Pesos — OR', color='#1D1D1F',
             fontsize=11, fontweight='bold', loc='left', pad=10)
ax.set_xlabel('Época', color=CINZA, fontsize=8.5)
ax.set_ylabel('Valor', color=CINZA, fontsize=8.5)
ax.set_xticks([1, 2, 3, 4])
ax.legend(frameon=False, labelcolor=CINZA, fontsize=8)

plt.tight_layout(pad=2.5)
plt.savefig('graficos_perceptron_OR.png', dpi=150,
            bbox_inches='tight', facecolor='white')
plt.show()
print('\n✅ Gráficos salvos como graficos_perceptron_OR.png')
```
